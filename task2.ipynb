```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
client_data_path = "https://raw.githubusercontent.com/jamesiswanto/BCG-X-Data-Science-Job-Simulation/refs/heads/development/client_data.csv"
price_data_path = "https://raw.githubusercontent.com/jamesiswanto/BCG-X-Data-Science-Job-Simulation/refs/heads/development/price_data.csv"

client_data = pd.read_csv(client_data_path)
price_data = pd.read_csv(price_data_path)
```

```python
# Data Types and Structure
print("Client Data Info:")
client_data.info()

print("\nPrice Data Info:")
price_data.info()
```

```python
# Descriptive Statistics
print("\nClient Data Descriptive Statistics:")
print(client_data.describe(include='all'))

print("\nPrice Data Descriptive Statistics:")
print(price_data.describe(include='all'))
```

```python
# Set up visualization style
sns.set(style="whitegrid")

# Visualizing Distributions in Client Data
numeric_cols_client = [
    "cons_12m", "cons_gas_12m", "forecast_cons_12m",
    "margin_net_pow_ele", "net_margin", "num_years_antig", "churn"
]

fig, axes = plt.subplots(len(numeric_cols_client), 1, figsize=(8, 30))
for i, col in enumerate(numeric_cols_client):
    sns.histplot(client_data[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Visualizing Distributions in Price Data
numeric_cols_price = [
    "price_off_peak_var", "price_peak_var", "price_mid_peak_var",
    "price_off_peak_fix", "price_peak_fix", "price_mid_peak_fix"
]

fig, axes = plt.subplots(len(numeric_cols_price), 1, figsize=(8, 25))
for i, col in enumerate(numeric_cols_price):
    sns.histplot(price_data[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
